# **0. Preliminary Concepts**

## **0.1 What is RAG?**

In Retrieval Augmented Generation (RAG), an LLM retrieves contextual documents from an external dataset as part of its execution.

This is useful if we want to ask question about specific documents (e.g., our PDFs, a set of videos, etc).

## **0.2. What is LangChain?**

LangChain is a framework for developing applications powered by language models. It provides a standard interface for interacting with different LLMs and tools, making it easier to build and deploy LLM applications. LangChain offers various components, including document loaders, text splitters, prompt templates, and chains, to facilitate the development process.

# **1. Loading Unstructured Data**

Document loaders deal with the specifics of accessing and converting data from a variety of different formats and sources into a standardized format. There are over 80 different types of document loaders in LangChain that can handle various data sources like websites, databases, YouTube videos, and different data types such as PDFs, HTML, and JSON.

The primary purpose of document loaders is to take this variety of data sources and load them into a standard document object, which consists of:
1. Content - The actual text or data
2. Metadata - Additional information about the document

Document loaders can be broadly categorized into:
- Loaders for unstructured data from public sources (YouTube, Twitter, etc.)
- Loaders for unstructured data from proprietary sources (Figma, Notion, etc.)
- Loaders for structured data in tabular formats (Airbyte, Stripe, Airtable, etc.)

## **1.1. PDFs**

PDFs are one of the most common document formats you'll want to work with. LangChain provides a `PyPDFLoader` to extract text from PDF files.

When loading a PDF, the loader will:
1. Read each page of the PDF
2. Convert each page into a separate Document object
3. Store the page number in the metadata
4. Extract the text content for each page

This allows you to later retrieve and work with specific pages based on their relevance to a query.

```python
if running_in_colab: 
    ! pip install pypdf
    from google.colab import drive
    drive.mount('/content/drive')
    folder_dir = '/content/drive/My Drive/'
    # In Colab, you should have an HF_TOKEN in Secrets
else: 
    from dotenv import load_dotenv, find_dotenv
    _ = load_dotenv(find_dotenv()) # read local .env file. You should add your own HF_TOKEN in the .env file
    folder_dir = 'docs/'
```

```python
from langchain.document_loaders import PyPDFLoader
loader = PyPDFLoader(folder_dir + 'paper.pdf')
pages = loader.load()
print(f"\n Number of pages in the document: {len(pages)}")
```

- Each page is a `Document`.
- A `Document` contains text (`page_content`) and `metadata`.

```python
print(f"{type(pages) = }")
print(f"{type(pages[0]) = }")
```

When we load a PDF, we get a list of Document objects - one for each page in the PDF. Each Document object has two key components:

1. **page_content**: The actual text content extracted from the page.
2. **metadata**: Additional information about the document, including:
   - **source**: The path to the PDF file
   - **page**: The page number within the PDF

This structure is consistent across different document loaders in LangChain, which allows for a standardized way to process and retrieve information regardless of the original source format.

> Note: We can access the content of each page using the `.page_content` attribute, and we can access the metadata using the `.metadata` attribute.

```python
# What is the content of each page?
for i, page in enumerate(pages):
    print(f"\nð Page {i+1}: Content")
    print(page.page_content)
```

```python
# What is the metadata of each page?
for i, page in enumerate(pages):
    print(f"\nð Page {i+1}: Metadata")
    print(page.metadata)
```

## **1.2. URLs**

The internet contains a vast amount of educational and useful content that you might want to incorporate into your RAG applications. LangChain's `WebBaseLoader` allows you to load content directly from URLs.

When using the WebBaseLoader:
1. The loader fetches the content from the specified URL
2. It extracts the text content from the HTML
3. It creates a Document object with the extracted content

Keep in mind that web content often contains a lot of formatting, navigation elements, and other non-essential information. You might notice white spaces and extra text in the loaded content. This is why post-processing is typically needed after loading web content to get it into a cleaner, more workable format.

```python
from langchain.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://github.com/basecamp/handbook/blob/master/titles-for-programmers.md")
```

```python
docs = loader.load()
```

```python
print(docs[0].page_content[:1000])
```

## **1.3. YouTube**

YouTube is a rich source of video content that can be valuable for many applications. LangChain provides tools to extract and work with information from YouTube videos.

Loading YouTube content involves two main components:
1. **YoutubeAudioLoader**: This loads the audio from a YouTube video
2. **FasterWhisperParser**: This uses a speech-to-text model to convert the audio into text format

The process works as follows:
1. The audio from the specified YouTube URL is downloaded
2. The audio is saved to the specified directory
3. The speech-to-text model transcribes the audio
4. The transcript is converted into Document objects

This allows you to ask questions about your favorite videos, lectures, or any other YouTube content. The transcription process might take a few minutes depending on the length of the video and the computational resources available.

```python
from langchain_community.document_loaders.generic import GenericLoader
from langchain_community.document_loaders.blob_loaders.youtube_audio import YoutubeAudioLoader
from langchain_community.document_loaders.parsers.audio import FasterWhisperParser
```

**Note**: This can take several minutes to complete.

```python
url="https://www.youtube.com/watch?v=lzILoMjEpaE" # "https://www.youtube.com/watch?v=J7b0jxVB1TE"
save_dir="output/2_3/youtube/"
import os
os.makedirs(save_dir, exist_ok=True)  # Ensure the directory exists

loader = GenericLoader(
    YoutubeAudioLoader([url],save_dir),  # fetch from youtube
    FasterWhisperParser(device="gpu", model_size="small")  # or model_size="base", "medium", "large-v3"
)
docs = loader.load()
```

```python
for i, doc in enumerate(docs): 
    doc.page_content = doc.page_content.replace("\n", " ")
    print(f"\nð Page {i}: {doc.page_content}")
```
